```python
data = pd.read_csv('labels.csv',sep='\t')
```

```python
df_clean = data.dropna(subset=['label'])
df_clean = df_clean.dropna(subset=['mask'])
df_clean
```

```python
df_clean['label'].unique()
```

```python
#Filter list elements starting with given prefix using list comprehension + startswith()
start_letter = 'c'

colonoscopy_labels = [x for x in df_clean['label'].unique() if x.startswith(start_letter)]
colonoscopy_labels
```

```python
#Filter list elements starting with given prefix using list comprehension + startswith()
uc_list = ['c32','c33','c34']

uc_labels = [x for x in df_clean['label'].unique() if x in uc_list]
uc_labels
```

```python
df_colonoscopy = df_clean.apply(lambda row: row[df_clean['label'].isin(colonoscopy_labels)])
df_colonoscopy
```

```python
df_uc = df_colonoscopy.apply(lambda row: row[df_colonoscopy['label'].isin(uc_labels)])
df_uc = df_uc[~df_uc['sequence'].isin(['samples'])]
df_uc
```

```python
df_no =  df_colonoscopy.apply(lambda row: row[~df_colonoscopy['label'].isin(uc_labels)])
df_no = df_no[~df_no['sequence'].isin(['samples'])]
df_no
```

```python
import os
os.makedirs("seq_0000/benign/", exist_ok=True)
os.makedirs("seq_0000/malignant/", exist_ok=True)
os.makedirs("seq_0000/benign_masked/", exist_ok=True)
os.makedirs("seq_0000/malignant_masked/", exist_ok=True)
```

```python
malignant_files = df_uc['file'].values
malignant_masks = df_uc['mask'].values
benign_files = df_no['file'].values
benign_masks = df_no['mask'].values

# Check whether we have equal masks as samples
assert(len(malignant_files) == len(malignant_masks))
assert(len(benign_files) == len(benign_masks))
def save_img(files, masks, label=0, type="benign"):
    n = len(files)
    for i, img_path in tqdm(enumerate(files)):
        try:
            path_args = img_path.split('/')
            img_folder, img_file = "/".join(path_args[:-1]), path_args[-1]
            save_folder = img_folder.replace("/", "-")
            frame = int(img_file.split('.png')[0])

            other_frames = [f'{j:06}.png' for j in range(frame-2, frame+3) if j!=frame ]

            for f_name in other_frames:
                resized = Image.open(f"{img_folder}/{f_name}").resize((720, 576))
                Image.fromarray(np.asarray(resized, dtype=np.uint8)).save(f"seq_0000/{type}/{save_folder}_{f_name}")
            
            #Read img and mask in and save masked image
            img = Image.open(files[i])
            mask = Image.open(masks[i])
            W, H = img.size
            W_m, H_m = mask.size
            img = np.asarray(img.resize((720, 576)), dtype=np.uint8)
            mask = np.asarray(mask.resize((720, 576)), dtype=np.bool_)
            mask = mask[:, :, :3]
            masked_img = img*mask
            im = Image.fromarray(masked_img)
            im.save(f'seq_0000/{type}_masked/{save_folder}_{frame}_masked.png')
        except Exception as e:
            continue
        #break
save_img(benign_files, benign_masks, label=0, type="benign")
save_img(malignant_files, malignant_masks, label=1, type="malignant")
```
